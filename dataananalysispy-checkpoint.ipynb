# EXPLORATORY DATA ANALYSIS

```python
df.columns
```

# Gender

```python
ax=sbn.countplot(x='Gender',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
gensale=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sbn.barplot(x='Gender',y='Amount',data=gensale)
```

From the above graphs we can conclude that purchasing power of female is more than men.

# Age

```python
df.columns
```

```python
bx=sbn.countplot(x='Age Group',data=df,hue='Gender')
for bars in bx.containers:
  bx.bar_label(bars)
```

```python
agesale=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sbn.barplot(x='Age Group',y='Amount',data=agesale)
```

From the above graph we can conclude that most of the buyers are from age group of between 26-35 

# State

```python
statesale=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sbn.set(rc={'figure.figsize':(17,5)})
sbn.barplot(x='State',y='Orders',data=statesale)
```

```python
statesale=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sbn.set(rc={'figure.figsize':(17,5)})
sbn.barplot(x='State',y='Amount',data=statesale)
```

From the above graph we can conclude that most of the ordes and highest sales amount are from Uttar Pradesh,Maharashtra,Karnataka and Delhi respectively.

# Marital Status

```python
sbn.set(rc={'figure.figsize':(7,5)})
bx=sbn.countplot(x='Marital_Status',data=df,hue='Gender')
for bars in bx.containers:
  bx.bar_label(bars)
```

# Occupation

```python
sbn.set(rc={'figure.figsize':(20,5)})
bx=sbn.countplot(x='Occupation',data=df)
for bars in bx.containers:
  bx.bar_label(bars)
```

```python
occsale=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sbn.set(rc={'figure.figsize':(17,5)})
sbn.barplot(x='Occupation',y='Amount',data=occsale)
```

From the above graph we can conclude that most of the orders are from Healthcare,government and automobile sector and highest sales amount are from IT sector,healthcare and aviation respectively. respectively.

# Product category

```python
sbn.set(rc={'figure.figsize':(23,5)})
bx=sbn.countplot(x='Product_Category',data=df)
for bars in bx.containers:
  bx.bar_label(bars)
```

```python
prosale=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sbn.set(rc={'figure.figsize':(17,5)})
sbn.barplot(x='Product_Category',y='Amount',data=prosale)
```

# Conclusion

Married women of age26-35 years from UP ,Maharshtra and Karnataka working in IT Sector,Healthcare and aviation are most likely to buy products from food,clothing and electronics categories.
